```python
x = docs.TEXT
y = docs.LABEL
print(x.shape)
print(y.shape)
```

```python
from sklearn.model_selection  import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y,test_size = 0.20, random_state=26)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vect = TfidfVectorizer()
```

```python
vect.fit(x_train)
vect.vocabulary_
```

```python
# removing the stop words
vect = TfidfVectorizer(stop_words='english')
```

```python
vectorizer = vect.fit(x)
```

```python
# transforming the train and test datasets
x_train_transformed = vect.transform(x_train)
x_test_transformed =vect.transform(x_test)
```

```python
print(type(x_train_transformed))
print(x_train_transformed)
```

```python
from sklearn.svm import SVC

model = SVC(kernel = 'linear', random_state = 10,verbose= True)
model.fit(x_train_transformed,y_train)


# predict class
y_pred_class = model.predict(x_test_transformed)
```

```python
from sklearn import metrics
print(metrics.accuracy_score(y_test, y_pred_class))
print(metrics.f1_score(y_test, y_pred_class, average='macro'))
print(metrics.precision_score(y_test, y_pred_class, average='macro'))
print(metrics.recall_score(y_test, y_pred_class, average='macro'))
```

```python
import nlpaug.augmenter.word as naw
```

```python
syn_aug = naw.SynonymAug(aug_p=0.5)
```

```python
text_dict = {'LABEL':[], 'TEXT':[]}
for index,row in docs.iterrows():
    aug_txt = syn_aug.augment(row['TEXT'], n=1)
    if aug_txt != row['TEXT']:
        text_dict['LABEL'].append(row['LABEL'])
        text_dict['TEXT'].append(aug_txt[0])
text_dict = pd.DataFrame(text_dict)
docs = pd.concat([docs, text_dict], ignore_index = True)
    
```

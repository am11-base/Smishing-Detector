```python
X=datasets['TEXT']
Y=datasets['LABEL']
```

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
stopwords.words('english')
import re
from sklearn.feature_extraction.text import TfidfVectorizer
import string
```

```python
stopwords = nltk.corpus.stopwords.words('english')
ps = nltk.WordNetLemmatizer()
```

```python
def clean_text(text):
    no_punctuation = ''.join([word.lower() for word in text if word not in string.punctuation])
    tokens = re.split('\W+', text)
    stems = [ps.lemmatize(word) for word in tokens if word not in stopwords] # Remove Stopwords
    return stems
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vect = TfidfVectorizer(analyzer=clean_text)
vector=vect.fit_transform(X)
X=pd.DataFrame(vector.toarray())
```

```python
from sklearn.model_selection  import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, Y,test_size = 0.20, random_state=26)
```

```python
import xgboost as xgb
classifier= xgb.XGBClassifier()


classifier.fit(x_train,y_train)
predicted = classifier.predict(x_test)
pred_prob=classifier.predict_proba(x_test)
print(metrics.accuracy_score(y_test, predicted))
print(metrics.roc_auc_score(y_test,pred_prob,multi_class='ovr'))
print(metrics.f1_score(y_test,predicted,average='micro'))
```

```python
from sklearn.ensemble import GradientBoostingClassifier

classifier= GradientBoostingClassifier()
classifier.fit(x_train,y_train)
predicted = classifier.predict(x_test)
pred_prob=classifier.predict_proba(x_test)
print(metrics.accuracy_score(y_test, predicted))
print(metrics.roc_auc_score(y_test,pred_prob,multi_class='ovr'))
print(metrics.f1_score(y_test,predicted,average='micro'))
```

```python
from sklearn.svm import SVC

classifier= SVC(kernel = 'linear', random_state = 10,verbose= True, probability=True)
classifier.fit(x_train,y_train)
predicted = classifier.predict(x_test)
pred_prob=classifier.predict_proba(x_test)
print(metrics.accuracy_score(y_test, predicted))
print(metrics.roc_auc_score(y_test,pred_prob,multi_class='ovr'))
print(metrics.f1_score(y_test,predicted,average='micro'))
```

```python
from sklearn.naive_bayes import MultinomialNB
mnb = MultinomialNB()

# fit
naive_model = mnb.fit(x_train,y_train)

# predict class
y_pred_class = mnb.predict(x_test)

# printing the overall accuracy
from sklearn import metrics
metrics.accuracy_score(y_test, y_pred_class)
```

```python
rf=RandomForestClassifier()
rus_model=rf.fit(x_train,y_train)
y_rus_class=rf.predict(x_test)
metrics.accuracy_score(y_test,y_rus_class)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
model = KNeighborsClassifier()
model.fit(x_train,y_train)

predictions = model.predict(x_test)
print("K-Nearest Neighbour = ",metrics.accuracy_score(y_test, predictions))
```

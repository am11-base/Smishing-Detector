# Random Oversampling

```python
from imblearn.over_sampling import RandomOverSampler
sampler=RandomOverSampler(sampling_strategy='not majority')
x_res,y_res=sampler.fit_resample(X.values.reshape(-1,1),Y)
print(sorted(Counter(y_res).items()),y_res.shape)
```

```python
print(type(x_res),type(y_res))
```

```python
df1=pd.DataFrame(x_res,columns=['TEXT'])
```

```python
print(x_res,y_res)
```

```python
result=pd.concat([df1,y_res],axis=1,join="inner")
```

```python
result.to_csv("E:/Users/hashim/Documents/cse 19-23/4th yr/project/dataset/gsheets/oversam_data1.csv",index=False,mode='w')
```
